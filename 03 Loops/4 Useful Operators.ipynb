## 1. range

The range function allows you to quickly *generate* a list of integers, this comes in handy a lot, so take note of how to use it! There are 3 parameters you can pass, a start, a stop, and a step size.
**<code> range(start,stop,step) </code>**

Let's see some examples:

```python
range(0,11)
```

Note that this is a **generator** function, so to actually get a list out of it, we need to cast it to a list with **list()**. 
What is a generator? Its a special type of function that will generate information and not need to save it to memory.

```python
# 11 will not be included
list(range(0,11))
```

```python
# till now we used just two of the three notations in range()
# lets see if we want to print number with a gap of some numbers, say 2

new_list=list(range(0,10,2))
print(new_list)
from IPython.display import Image

Image("../img/range.png")
```

## 2. enumerate

enumerate is a very useful function to use with for loops.
in other words,
<code>Enumerate()</code> method adds a counter to an iterable and returns it in a form of enumerate object. This enumerate object can then be used directly in <code>for</code> loops or be converted into a list of <code>tuples</code> using <code>list()<code> method.

```python
index_count = 0

for letter in 'abcde':
    print("At index {} the letter is {}".format(index_count,letter))
    index_count += 1
```

Keeping track of how many loops you've gone through and where you are is very common and important. <code>enumerate()</code> was created, so developers wouldn't need to worry about creating and updating the index_count or loop_count.

Lets take a closer look:

```python
for i,letter in enumerate('abcde'):
    print('At index {} the letter is {}'.format(i,letter))
```

Now I prefer to achieve as many things as possible with as little code as possible. <code>enumerate()</code> helps.

## 3. zip

Notice the format enumerate actually returns, let's take a look by transforming it to a list()

```python
list(enumerate('abcde'))
```

```python
# Lets see what happens when we try to zip two lists together
mylist1 = [1,2,3,4,5]
mylist2 = ['a','b','c','d','e']

list(zip(mylist1,mylist2))
```

```python
# or if you prefer to write a for loop:

for item1,item2 in zip(mylist1,mylist2):
    print('For this tuple, item1 is: {} and item2 is: {}'.format(item1,item2))
```

## 4. in operator

We've already seen the **in** keyword during the for loop, but we can also use it to quickly check if an object is in a list

```python
'x' in ['x','t','120']
```

## 5. min and max

Quickly check the minimum or maximum of a list with these functions.

```python
mylist = [1,2,4,6,7,3,9,100]

print(f'Min is: {min(mylist)}')
print(f'Max is: {max(mylist)}')
```
